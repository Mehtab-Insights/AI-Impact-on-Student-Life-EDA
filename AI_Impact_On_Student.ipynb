# EDA

### Age Distribution

```python
plt.hist(df['Age'], bins=15, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.tight_layout()
plt.show()
```

### Daily Usage Hours Distribution

```python
sns.histplot(data=df,x='Daily_Usage_Hours', bins=15,edgecolor='black')
plt.xlabel('Daily Usage Hours')
plt.ylabel('Count')
plt.title('Daily Usage Hours Distribution')

plt.tight_layout()
plt.show()
```

### Gender Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

gender_counts = df['Gender'].value_counts()
axes[0].bar(gender_counts.index, gender_counts.values, color=['pink', 'lightblue'], edgecolor='black')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Count')
axes[0].set_title('Gender Distribution')

# Pie chart
axes[1].pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', colors=['pink', 'lightblue'])
axes[1].set_title('Gender Percentage')

plt.tight_layout()
plt.show()

print("Gender Distribution:")
print(gender_counts)
```

### Education Level Distribution

```python
plt.subplots(figsize=(10, 5))

edu_counts = df['Education_Level'].value_counts()

plt.barh(edu_counts.index, edu_counts.values, color='lightgreen', edgecolor='black')
plt.xlabel('Count')
plt.title('Education Level Distribution')

plt.tight_layout()
plt.show()

print("Education Level Distribution:")
print(edu_counts)
```

### AI Tool Used Distribution

```python
plt.subplots(figsize=(10, 5))

tool_counts = df['AI_Tool_Used'].value_counts()
plt.barh(tool_counts.index, tool_counts.values, color='lightyellow', edgecolor='black')
plt.xlabel('Count')
plt.title('AI Tool Used Distribution')

plt.tight_layout()
plt.show()

print("AI Tools Used:")
print(tool_counts)
```

### Purpose of AI Usage

```python
plt.subplots(figsize=(10, 5))

purpose_counts = df['Purpose'].value_counts()
plt.barh(purpose_counts.index, purpose_counts.values, color='lightsteelblue', edgecolor='black')
plt.xlabel('Count')
plt.title('Purpose of AI Tool Usage')

plt.tight_layout()
plt.show()

print("Purpose Distribution:")
print(purpose_counts)
```

### Impact on Grades

```python
plt.subplots(figsize=(10, 5))

impact_counts = df['Impact_on_Grades'].value_counts()
colors_impact = ['#FF6B6B', '#4ECDC4', '#95E1D3']
plt.barh(impact_counts.index, impact_counts.values, color=colors_impact[:len(impact_counts)], edgecolor='black')
plt.xlabel('Count')
plt.title('Impact of AI on Grades')

plt.tight_layout()
plt.show()

print("Impact on Grades Distribution:")
print(impact_counts)
```

### Satisfaction Level

```python
fig, ax = plt.subplots(figsize=(10, 5))

satisfaction_counts = df['Satisfaction_Level'].value_counts()
colors_sat = ['#FFD93D', '#6BCB77', '#4D96FF']
plt.bar(satisfaction_counts.index, satisfaction_counts.values, color=colors_sat[:len(colors_sat)], edgecolor='black')
plt.ylabel('Count')
plt.title('Student Satisfaction Level with AI Tools')

plt.tight_layout()
plt.show()

print("Satisfaction Level Distribution:")
print(satisfaction_counts)
```

### City Distribution

```python
plt.subplots(figsize=(10, 5))

city_counts = df['City'].value_counts()
plt.barh(city_counts.index, city_counts.values, color='plum', edgecolor='black')
plt.xlabel('Count')
plt.title('Students by City')

plt.tight_layout()
plt.show()

print("City Distribution:")
print(city_counts)
```

### Gender vs AI Tools Used

```python
plt.subplots(figsize=(12, 5))

sns.countplot(x='Gender',hue='AI_Tool_Used',data=df)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('AI Tools Used by Gender')


plt.tight_layout()
plt.show()
```

### Impact on Grades vs Satisfaction

```python
plt.subplots(figsize=(10, 5))

sns.countplot(x='Impact_on_Grades',hue='Satisfaction_Level',data=df)
plt.xlabel('Impact on Grades')
plt.ylabel('Count')
plt.title('Impact on Grades vs Satisfaction Level')
plt.legend(title='Satisfaction', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Purpose vs Impact on Grades

```python
plt.subplots(figsize=(10, 5))

sns.countplot(x='Purpose',hue='Impact_on_Grades',data=df)
plt.xlabel('Purpose of Use')
plt.ylabel('Count')
plt.title('Purpose of AI Usage vs Impact on Grades')

plt.legend(title='Impact', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# Lastly Compact interactive dashboard that summarizes the key EDA charts

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Prepare data
age_data = df['Age']
usage_data = df['Daily_Usage_Hours']
gender_counts = df['Gender'].value_counts()
edu_counts = df['Education_Level'].value_counts()
tool_counts = df['AI_Tool_Used'].value_counts()
purpose_counts = df['Purpose'].value_counts()
impact_counts = df['Impact_on_Grades'].value_counts()
sat_counts = df['Satisfaction_Level'].value_counts()

# Color palette
colors = ['#4E79A7', '#F28E2B', '#59A14F', '#E15759', '#76B7B2', '#EDC948', '#B07AA1', '#FF9DA7']

fig = make_subplots(
    rows=2,
    cols=4,
    subplot_titles=(
        'Age Distribution',
        'Daily Usage Hours',
        'Gender Distribution',
        'Education Level',
        'AI Tool Used',
        'Purpose of AI Usage',
        'Impact on Grades',
        'Satisfaction Level'
    ),
    specs=[
        [{"type": "xy"}, {"type": "xy"}, {"type": "domain"}, {"type": "xy"}],
        [{"type": "xy"}, {"type": "xy"}, {"type": "xy"}, {"type": "xy"}]
    ],
    vertical_spacing=0.18,
    horizontal_spacing=0.08
)

# 1) Age histogram (row 1, col 1)
fig.add_trace(
    go.Histogram(x=age_data, nbinsx=15, name='Age', marker_color=colors[0], showlegend=False,
                 hovertemplate='Age: %{x}<br>Count: %{y}<extra></extra>'),
    row=1, col=1
)

# 2) Daily Usage Hours histogram (row 1, col 2)
fig.add_trace(
    go.Histogram(x=usage_data, nbinsx=15, name='Daily Usage', marker_color=colors[1], showlegend=False,
                 hovertemplate='Hours: %{x}<br>Count: %{y}<extra></extra>'),
    row=1, col=2
)

# 3) Gender pie chart (row 1, col 3)
fig.add_trace(
    go.Pie(labels=gender_counts.index, values=gender_counts.values, name='Gender',
           marker_colors=colors, hole=0.35, showlegend=True,
           hovertemplate='%{label}: %{value} (%{percent})<extra></extra>'),
    row=1, col=3
)

# 4) Education Level bar chart (row 1, col 4)
fig.add_trace(
    go.Bar(x=edu_counts.values, y=edu_counts.index, orientation='h', name='Education',
           marker_color=colors[2], showlegend=False,
           hovertemplate='%{y}: %{x}<extra></extra>'),
    row=1, col=4
)

# 5) AI Tool Used bar chart (row 2, col 1)
fig.add_trace(
    go.Bar(x=tool_counts.values, y=tool_counts.index, orientation='h', name='AI Tool',
           marker_color=colors[3], showlegend=False,
           hovertemplate='%{y}: %{x}<extra></extra>'),
    row=2, col=1
)

# 6) Purpose bar chart (row 2, col 2)
fig.add_trace(
    go.Bar(x=purpose_counts.values, y=purpose_counts.index, orientation='h', name='Purpose',
           marker_color=colors[4], showlegend=False,
           hovertemplate='%{y}: %{x}<extra></extra>'),
    row=2, col=2
)

# 7) Impact on Grades bar chart (row 2, col 3)
fig.add_trace(
    go.Bar(x=impact_counts.index, y=impact_counts.values, name='Impact',
           marker_color=colors[5], showlegend=False,
           hovertemplate='%{x}: %{y}<extra></extra>'),
    row=2, col=3
)

# 8) Satisfaction Level bar chart (row 2, col 4)
fig.add_trace(
    go.Bar(x=sat_counts.index, y=sat_counts.values, name='Satisfaction',
           marker_color=colors[6], showlegend=False,
           hovertemplate='%{x}: %{y}<extra></extra>'),
    row=2, col=4
)

fig.update_layout(
    title=dict(text='AI Impact on Student Life — EDA Summary Dashboard', font=dict(size=18)),
    template='plotly_white',
    height=750,
    width=1400,
    bargap=0.2,
    legend=dict(orientation='h', yanchor='bottom', y=1.05, xanchor='center', x=0.5)
)

# Axis labels
fig.update_xaxes(title_text='Age', row=1, col=1)
fig.update_yaxes(title_text='Count', row=1, col=1)
fig.update_xaxes(title_text='Hours/Day', row=1, col=2)
fig.update_yaxes(title_text='Count', row=1, col=2)
fig.update_xaxes(title_text='Count', row=1, col=4)
fig.update_xaxes(title_text='Count', row=2, col=1)
fig.update_xaxes(title_text='Count', row=2, col=2)
fig.update_yaxes(title_text='Count', row=2, col=3)
fig.update_yaxes(title_text='Count', row=2, col=4)

fig.write_html('dashboard.html')
fig.show()
```
